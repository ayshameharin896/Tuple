```python
#coversion of tuple to list and wiseversa
t = (23,53,22,456,43,76,43,5,53,76,43,54,34)
l = list(t)
t2 = tuple(l)
print(t2)
```

```python
#combining 2 tuple
t = (23,53,22,456,43,76,43,5,53,76,43,54,34)
t1=(32,33,5,64,3,87)
c=t+t1
print(c)
```

```python
#extracting the elements from 1 to 3 and 9th to 12th from a tuple
t = (23,53,22,456,43,76,43,5,53,76,43,54,34)
first=t[0:3]
last=t[9:12]
print(first)
print(last)
```

```python
#finding the sum of elements in a tuple
t = (23,53,22,456,43,76,43,5,53,76,43,54,34)
sum=sum(t)
print(sum)
```

```python
#adding an element in a tuple
a=('a','b','c','d','g')
list1=list(a)
list1[1]='k'
a=tuple(list1)
print(a)
```

```python
#storing string value in a tuple
a=("sd","ef","sd","wed","ase")
print(a)
```

```python
#extracting the last  value in a tuple
a=("sd","ef","sd","wed","ase")
print(a[0])
print(a[-1])
```

```python
#checking if an element is in tuple
a=("sd","ef","sd","wed","ase")
if "sd" in a:
  print("yes")
else:
  print("no")
```

```python
#counting the number of times an elemnt is repeated in tuple
a=("sd","ef","sd","wed","ase","sd")
b=a.count("sd")
print(b)
```

```python
#finding the maximum and minimum in a tuple
a=(2,34,21,43,23,54)
print(min(a))
print(max(a))
```

```python
#printing the elements in a tuple in the form of a list
a=(2,34,21,43,23,54)
for i in a:
  print(i)
```
